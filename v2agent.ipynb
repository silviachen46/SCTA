```python
# Import necessary libraries
import scanpy as sc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import helper functions from utils_agent
from utils_agent import (
    load_and_annotate_h5_files, filter_cells, normalize_log_transform,
    filter_lowqc_cells, save_high_var_gene, pca_and_plot_umap, 
    annoatate_by_markers, display_umap, 
    plot_cluster_distribution_by_BL, plot_cluster_frequencies,
    get_deg, get_enrichment_analysis
)

# Import marker genes for annotation
from markers import marker_genes

# Step 1: Load and Annotate Data
adata = load_and_annotate_h5_files()

# Step 2: Quality Control - Filter cells based on number of genes and counts
filter_cells(adata, min_cells=3, min_genes=200, max_genes=50000)

# Step 3: Normalize the data and log-transform
normalize_log_transform(adata, target_sum=1e4)

# Step 4: Filter out low-quality cells by mitochondrial content
adata = filter_lowqc_cells(adata, pct_counts_mt_upbound=10)

# Step 5: Identify and save highly variable genes
adata = save_high_var_gene(adata, n_top_genes=4000)

# Step 6: Perform PCA, compute UMAP, and plot it by cell type and leiden cluster
pca_and_plot_umap(adata, n_pcs=40, n_neighbors=10, resolution=0.5)

# Annotate cells using marker genes
annoatate_by_markers(adata, marker_genes)

# Display UMAP by cell type
display_umap(adata, target="cell_type")

# Display UMAP by leiden cluster
display_umap(adata, target="leiden")

# Plot cluster distribution by some metadata (for example, Batch/BL Number or Sample groups)
plot_cluster_distribution_by_BL(adata, target="cell_type")
plot_cluster_distribution_by_BL(adata, target="leiden")

# Optional: Plot cluster frequencies across groups (e.g., by sample group)
plot_cluster_frequencies(
    adata,
    cluster_key="leiden",
    group_key="condition",  # Replace with "sample_group" or other grouping if more descriptive
    sample_key="BL_number"
)

# Step 7: Perform Differentially Expressed Gene Analysis (DEG)
# Use leiden cluster for DEG or group by "condition" (e.g., experimental condition or treatment group)
get_deg(adata, top_n=20, groupby="condition", reference="Control")

# Step 8: Perform Gene Set Enrichment Analysis (GSEA)
get_enrichment_analysis(
    adata,
    target="CP",  # Choose between CP, BP, KEGG, or HALLMARK based on the analysis
    reference="Control",
    groupby="condition",
    file_path="/Users/silviachen/Documents/Software/SCAagent/h5_file/h.all.v2024.1.Hs.symbols.gmt",  # Predefined GSEA file
    top_n=10
)
```
